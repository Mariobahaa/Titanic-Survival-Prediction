```
import pandas as pd
import tensorflow
import keras
import numpy as np
import sklearn
```

```
import math

train = pd.read_csv("train.csv") 
test  = pd.read_csv("test.csv")
#'Name', 'Ticket', 'Cabin'
del train['PassengerId']
del train['Name']
del train['Ticket']
del train['Cabin']

del test['PassengerId']
del test['Name']
del test['Ticket']
del test['Cabin']

y= train.iloc[:,0]
X_temp= train.iloc[:,1:]

X_test = test.iloc[:,:]

y= np.array(y)
X_temp= np.array(X_temp)
X_test= np.array(X_test)

#Pclass, Sex, Age, SibSp, Parch, Fare, Embarked


Pclass = np.zeros([X_temp[:,0].size,3])
for i in range (0,X_temp[:,0].size):
  Pclass[i,X_temp[i,0]-1] = 1

Sex = np.zeros([X_temp[:,1].size])
for i in range(0, X_temp[:,0].size):
  if X_temp[i,1] == 'male' :
    Sex[i] = 1
  elif X_temp[i,1] == 'female':
    Sex[i] = 0

Sex   = Sex.reshape(Sex.size,1)
Age   = X_temp[:,2]
maxAge = max(Age)
minAge = min(Age)

s = 0
cnt = 0
for i in range(0,Age.size):
  if math.isnan(float(Age[i])) == False :
    s+= Age[i] 
    cnt+=1
miss = s/cnt
print(miss)
for i in range(0,Age.size):
  if math.isnan(float(Age[i])):
    Age[i] = miss

    
Age   = (Age-min(Age)) /(max(Age)-min(Age))
Age   = Age.reshape(Age.size,1)
SibSp = X_temp[:,3]
maxSibSp = max(SibSp)
minSibSp = min(SibSp)
SibSp = (SibSp-min(SibSp)) /(max(SibSp)-min(SibSp))
SibSp = SibSp.reshape(SibSp.size,1)
Parch = X_temp[:,4]
maxParch = max(Parch)
minParch = min(Parch)
Parch = (Parch-min(Parch)) /(max(Parch)-min(Parch))
Parch = Parch.reshape(Parch.size,1)
Fare  = X_temp[:,5]
maxFare = max(Fare)
minFare = min(Fare)
Fare  = (Fare-min(Fare)) /(max(Fare)-min(Fare))
Fare  = Fare.reshape(Fare.size,1)

Embarked = np.zeros([X_temp[:,6].size,3])
for i in range (0,X_temp[:,6].size):
  if X_temp[i,6] == "Q":
    Embarked[i,0] = 1
  elif X_temp[i,6] == "S":
    Embarked[i,1] = 1
  elif X_temp[i,6] == "C":
    Embarked[i,2] = 1

X= np.append(Pclass,Sex, axis=1)
X= np.append(X, Age, axis=1)
X= np.append(X, SibSp, axis=1)
X= np.append(X, Parch, axis=1)
X= np.append(X, Fare, axis=1 )
X= np.append(X, Embarked, axis=1)

print(X.shape)
print(X[0])
```

```
Pclass = np.zeros([X_test[:,0].size,3])
for i in range (0,X_test[:,0].size):
  Pclass[i,X_test[i,0]-1] = 1

Sex = np.zeros([X_test[:,1].size])
for i in range(0, X_test[:,0].size):
  if X_test[i,1] == 'male' :
    Sex[i] = 1
  elif X_test[i,1] == 'female':
    Sex[i] = 0

Sex   = Sex.reshape(Sex.size,1)
Age   = X_test[:,2]
Age   = (Age-minAge) /(maxAge-minAge)
Age   = Age.reshape(Age.size,1)
SibSp = X_test[:,3]
SibSp   = (SibSp-minSibSp) /(maxSibSp-minSibSp)
SibSp = SibSp.reshape(SibSp.size,1)
Parch = X_test[:,4]
Parch = (Parch-minParch) /(maxParch-minParch)
Parch = Parch.reshape(Parch.size,1)
Fare  = X_test[:,5]
Fare  = (Fare-minFare) /(maxFare-minFare)
Fare  = Fare.reshape(Fare.size,1)

Embarked = np.zeros([X_test[:,6].size,3])
for i in range (0,X_test[:,6].size):
  if X_test[i,6] == "Q":
    Embarked[i,0] = 1
  elif X_test[i,6] == "S":
    Embarked[i,1] = 1
  elif X_test[i,6] == "C":
    Embarked[i,2] = 1

X_testx = np.array([])
X_testx= np.append(Pclass,Sex, axis=1)
X_testx= np.append(X_testx, Age, axis=1)
X_testx= np.append(X_testx, SibSp, axis=1)
X_testx= np.append(X_testx, Parch, axis=1)
X_testx= np.append(X_testx, Fare, axis=1 )
X_testx= np.append(X_testx, Embarked, axis=1)

print(X_testx.shape)
print(X_testx[0])
```

```
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten
from keras.callbacks import ModelCheckpoint
from keras import losses
from keras.optimizers import Adam
from keras.models import Model
from keras import regularizers

checkpointer = ModelCheckpoint(filepath = 'T1.weights.best.hdf5', verbose=1, save_best_only = True)

model= Sequential()
model.add(Dense(32, activation='relu', input_dim=11, )) #kernel_regularizer=

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))

opt = Adam(lr=0.001)
model.compile(loss= 'binary_crossentropy' ,
              optimizer= opt,
              metrics=['accuracy'])

#model.summary()

model.fit(X,y,epochs=200,verbose=1,callbacks=[checkpointer], shuffle= True, validation_split=0.15, batch_size=8)

model.save('T1.h5')
```

```
model.load_weights('T1.weights.best.hdf5')
model.fit(X,y,epochs=100,verbose=1,callbacks=[checkpointer], shuffle= True, validation_split=0.15, batch_size=8)
```

```
#predsx = model.predict_classes(X_testx)
model.load_weights('T1.weights.best.hdf5')
model.evaluate(X,y)

p1 = model.predict_classes(X_testx)
```

# CSV Creation

```
#Voting Function
preds = p1

for i in range(0,p1.size):
  zero = 0
  one  = 0
  if p1[i] == 1:
    one+= 1
  else :
    zero += 1
  if p2[i] == 1:
    one+= 1
  else :
    zero += 1
  if p3[i] == 1:
    one+= 1
  else :
    zero += 1
  if p4[i] == 1:
    one+= 1
  else :
    zero += 1
  if p5[i] == 1:
    one+= 1
  else:
    zero += 1
  if p6[i] == 1:
    one+= 1
  else:
    zero += 1
  if p7[i] == 1:
    one+= 1
  else:
    zero += 1
  if one > zero:
    preds[i] = 1
  else:
    preds[i] = 0
```
